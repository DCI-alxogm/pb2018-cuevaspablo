```python
#Programa creado por José Pablo Cuevas Cázares
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
#(d2o/dt2)+(w2)*(o)=0
#(do/dt)=v
#(d2o/dt2)=dv/dt
#(dv/dt)=-(w2)o=f1
#(do/dt)=v=f2
#v(ti+1)=v(ti)+h*f1
#o(ti+1)=o(ti)+h*f2
e=100000
T=1
pi=3.14159265358979323846
h=0.00005
w=2*pi/T
o=np.linspace(180,180,e)
v=np.linspace(0,10,e)
t=np.linspace(0,10,e)
#En este apartado definí los arreglos necesarios para poder realizar las operaciones.
#Definí al aángulo como "o", esto para poder inicializarlo como un arreglo en donde siempre es 180 y que termine en 180, para que siempre sea periódico.
#
def fx(w,o):
    z=(-(pow(w,2)))*o
    return z
z=fx(w,o)
#Después definí una función para poder calcular el parámetro de la velocidad angular u omega, dándole un valor a dicha función
#como una variable de tipo z
```

```python
for i in range (e-1):
    v[i+1]=v[i]+(h*(fx(w,o[i])))
    o[i+1]=o[i]+(h*v[i])
#En este apartado escribí las ecuaciones proporcionadas en clase, las cuales cambian en un rango una unidad menor al parámetro "e"
#como después de cada iteración debe cambiar la velocidad y el ángulo, se le suma una unidad al arreglo de las velocidades y los ángulos
#y desués utilizo la función previamente definida para poder calcular los valores correctos.
```

```python
plt.plot(v,o)
#En esta sección grafiqué los datos de la velocidad y los ángulos, lo cual nos proporciona una elipse, que es lo esperado, puesto que estos dos valores
#cambian de manera periódica, por tanto completando un ciclo, que graficado se observa como se ve en la siguiente imagen
```
